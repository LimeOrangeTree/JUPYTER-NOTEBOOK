# LOOP VERSION
- 게시판의 제목 및 게시글의 링크 수집

```python
from time import sleep
from random import randint
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36',
    'referer' : 'http://www.naver.com'
    }
```

```python
url1 = 'https://www.hibrain.net/research/researches/34/recruitments/108/categories/MJR/categories/ALL/recruits?page='
url2 = '&pagingno=1&listType=ING&pagesize=10&sortType=AD&cntType=MJR&limit=25&displayType=TIT&siteid=1'
```

```python
page = ['1','2','3']
```

```python
targetUrl = [] # 페이지 url 생성
for i in page:
    tmp = url1 + i + url2
    targetUrl.append(tmp)
```

```python
res = requests.get(targetUrl[0]) # http 요청
soup = BeautifulSoup(res.text) # page source를 beautifulsoup 객체에 넣기
Info = soup.find_all('li',{'class':'row sortRoot'}) # 요소 인덱싱
```

```python
hibrain = 'https://www.hibrain.net'
POST = pd.DataFrame(columns=['title','link','date','View'])

for url in targetUrl:
    res = requests.get(url, headers = headers)
    soup = BeautifulSoup(res.text)
    Info = soup.find_all('li',{'class':'row sortRoot'})
    for i in range(0,len(Info)):
        tmpDict = {'title' : soup.find_all('li',{'class':'row sortRoot'})[i].a['title'],
              'link' : hibrain+soup.find_all('li',{'class':'row sortRoot'})[i].a['href'],
              'date' : soup.find_all('li',{'class':'row sortRoot'})[i].find('span',{'class':'td_rdtm number'}).text,
              'View' : soup.find_all('li',{'class':'row sortRoot'})[i].find('span',{'class':'td_cnt number last'}).text}
        TMP = pd.DataFrame.from_dict(tmpDict, orient='index').transpose()
        POST = POST.append(TMP)
    sleep(randint(1,3))
```
