# 클리앙 게시판 데이터 수집 - 단일 페이지 + 게시글 제목

```python
from bs4 import BeautifulSoup
import requests # python HTTP library

tmpUrl = 'https://www.clien.net/service/group/allinfo' # 클리앙 정보와 자료 게시판
r = requests.get(tmpUrl)
print(r) # <Response [200]> : 성공적으로 연결
```

```python
html =r.text # url의 page source 가져오기
soup = BeautifulSoup(html, 'html.parser')
```

```python
PostList = soup.find_all('span', {'class':'subject_fixed'}) # tag, {'class' : 'class name'} - 게시물 제목 인덱싱
```

```python
print(len(PostList))
```

```python
PostList
```

```python
TmpPostList = []
for post in PostList:
    print(post.text)
    tmp = post.text
    TmpPostList.append(tmp)
```

```python
TmpPostList
```

```python
# post view indexing
PostView = soup.find_all('span', {'class':'hit'})
TmpViewList = []
for view in PostView:
    TmpViewList.append(view.text)
    print(view.text)
```

```python
TmpViewList
```

```python
# post date indexing
PostDate = soup.find_all('span', {'class':'timestamp'})
TmpDateList = []
for date in PostDate:
    TmpDateList.append(date.text)
    print(date.text)
```

```python
print(len(TmpPostList))
print(len(TmpDateList))
print(len(TmpViewList))
```

```python
TmpDateList
```

```python
TmpDateList[1:]
```

```python
print(len(TmpDateList[1:]))
TmpDateList= TmpDateList[1:]
```

```python
TmpViewList
```

```python
TmpViewList[1:]
```

```python
TmpViewList= TmpViewList[1:]
```

```python
print(len(TmpPostList))
print(len(TmpDateList))
print(len(TmpViewList))
```

```python
import pandas as pd

column = ['Post_Name', 'Post_View', 'Post_Date']
TMPDF = pd.DataFrame(columns=column)
TMPDF
```

```python
TMPDF['Post_Name'] = TmpPostList
TMPDF['Post_Date'] = TmpDateList
TMPDF['Post_View'] = TmpViewList
```

# 클리앙 게시판 데이터 수집 - 단일 페이지 + 게시글 제목 + 게시글 내용

```python
from bs4 import BeautifulSoup
import requests # python HTTP library

tmpUrl = 'https://www.clien.net/service/group/allinfo' # 클리앙 정보와 자료 게시판
r = requests.get(tmpUrl)
print(r) # <Response [200]> : 성공적으로 연결

html =r.text # url의 page source 가져오기
soup = BeautifulSoup(html, 'html.parser')
```

```python
postList = soup.find_all('a', {'class':'list_subject'})
postList
```

```python
for post in postList:
    print(post['href'])
```

```python
clienUrl = 'https://www.clien.net'

PostUrl = []
for post in postList:
    tmpUrl = clienUrl + post['href']
    print(tmpUrl)
    PostUrl.append(tmpUrl)
```

```python
print(len(PostUrl)) # list의 요소 갯수  -> 갯구확인
PostUrl # 게시물 url list
```

```python
PostUrl = PostUrl[1:]
```

```python
# 1가지 url만 사용하여 post content 추출
print(PostUrl[0])
response = requests.get(PostUrl[0])
html = response.text
soup = BeautifulSoup(html, 'html.parser')
soup.find('div', {'class':'post_content'}).text # post의 content 추출
```

```python
import time

TmpPostContent = []
for url in PostUrl:    
    print('Input Url:', url)
    response = requests.get(url)
    html = response.text
    soup = BeautifulSoup(html, 'html.parser')
    tmpContent = soup.find('div', {'class':'post_content'}).text # post의 content 추출
#     print(tmpContent)
    TmpPostContent.append(tmpContent)
    time.sleep(1) # 요청 사이에 1초 휴식!
```

```python
print(len(TmpPostContent))
TMPDF['Post_Content'] = TmpPostContent
```

```python
TMPDF
```

```python
# dataframe column reindex
column = ['Post_Name','Post_Content','Post_View','Post_Date']
TMPDF.reindex(column, axis = 'columns')
```
