# Python을 이용하여 이미지 다운로드
- urllib.request.urlretrieve : url를 local file로 다운로드 받을 수 있는 함수

### 네이버 이미지 다운로드 - BeautifulSoup version

```python
from selenium import webdriver
from bs4 import BeautifulSoup
from random import randint
from time import sleep
from tqdm import tqdm
import requests
import urllib.request
```

```python
targetUrl = 'https://www.naver.com/'
```

```python
res = requests.get(targetUrl)
soup = BeautifulSoup(res.text)
```

```python
soup
```

```python
ImageUrl = []
for url in soup.find_all('img'): # parsing image url 
    if 'https' in url['src']:
        print(url['src'])
        ImageUrl.append(url['src'])
```

```python
requests.get(ImageUrl[0], stream=True)
```

```python
ImageUrl
```

```python
print(len(ImageUrl))
```

```python
saveDir = '/Users/jiwonjang/python_tutorial/ve_env/강의자료_크롤링/image_download/'
urllib.request.urlretrieve(ImageUrl[5],saveDir+str(1)+'.jpg') # image download
```

```python
saveDir = '/Users/jiwonjang/python_tutorial/ve_env/강의자료_크롤링/image_download/'
for z, url in enumerate(ImageUrl):
    sleep(randint(1,2))
    urllib.request.urlretrieve(url,saveDir+str(z)+'.jpg')
```

### 인스타그램 이미지 다운로드 - Selenium version

```python
from selenium.webdriver.common.keys import Keys 
```

```python
targetUrl = 'https://www.instagram.com/explore/tags/%ED%95%98%EB%A6%BC%ED%8E%AB%ED%91%B8%EB%93%9C'
```

```python
chromeDir = '/Users/jiwonjang/python_tutorial/chromedriver'
driver = webdriver.Chrome(chromeDir)
driver.get(targetUrl)
```

```python
driver.implicitly_wait(3)
```

```python
images = driver.find_elements_by_tag_name('img')
```

```python
print(len(images))
```

```python
ImageList = []
for i in images:
    tmp = i.get_attribute('src')
    ImageList.append(tmp)
```

```python
ImageList
```

```python
print(len(ImageList))
```

```python
saveDir = '/Users/jiwonjang/python_tutorial/ve_env/강의자료_크롤링/image_download/'
for z, url in enumerate(ImageList):
    sleep(randint(1,2))
    urllib.request.urlretrieve(url,saveDir+str(z)+'.jpg')
```

# CLIEN 게시판
- 게시글에서 이미지만 저장하고 싶은 경우

```python
clienUrl = 'https://www.clien.net/'
targetUrl = 'https://www.clien.net/service/group/allinfo?&od=T31&po=0'
```

```python
res = requests.get(targetUrl)
soup = BeautifulSoup(res.text)
```

```python
PostUrls = []
for i in soup.find_all('a', {'class':'list_subject'}): # link list 저장
    PostUrls.append(i['href'])
    print(i['href'])
```

```python
FinalPosturl = []
for url in PostUrls: # 절대경로 만들기
    print(clienUrl+url)
    FinalPosturl.append(clienUrl+url)
```

```python
FinalPosturl[1:]
```

```python
res = requests.get(FinalPosturl[1])
soup = BeautifulSoup(res.text)
contents = soup.find('div',{'class':'post_article fr-view'})
```

```python
imageUrl = contents.p
```

```python
imageUrl.find('img')['src']
```

```python
if 'https' in imageUrl.find('img')['src']:
    print(imageUrl.find('img')['src'])
    
```

### loop version

```python
# parse image url
ImageUrls = []
for url in tqdm(FinalPosturl[1:]):
    sleep(randint(1,2))
    res = requests.get(url)
    soup = BeautifulSoup(res.text)
    contents = soup.find('div',{'class':'post_article fr-view'})
    imageUrl = contents.p
    print('Parsed tag:', imageUrl)
    if 'https' in imageUrl.find('img')['src']:
        print(imageUrl.find('img')['src'])
        image = imageUrl.find('img')['src']
        ImageUrls.append(image)
```

------------------------------------------------------------

```python
# parse image url with exception
ImageUrls = []
for url in tqdm(FinalPosturl[1:]):
    sleep(randint(1,2))
    res = requests.get(url)
    soup = BeautifulSoup(res.text)
    contents = soup.find('div',{'class':'post_article fr-view'})
    imageUrl = contents.p
    print('Parsed tag:', imageUrl)
    try:
        if 'https' in imageUrl.find('img')['src']:
            print(imageUrl.find('img')['src'])
            image = imageUrl.find('img')['src']
            ImageUrls.append(image)
    except:
        pass
```

### download images

```python
# download image
saveDir = '/Users/jiwonjang/python_tutorial/ve_env/강의자료_크롤링/image_download/'
for z, url in tqdm(enumerate(ImageUrls)):
    sleep(randint(1,2))
    urllib.request.urlretrieve(url,saveDir+str(z)+'.jpg')
print('downloads finish!')
```
