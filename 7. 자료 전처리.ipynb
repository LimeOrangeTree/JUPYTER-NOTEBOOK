### loop version

```python
ViewCount = []
for i in CLIEN['View']:
    if 'k' in i:
        tmp = i.replace('k','')
        tmp = tmp.strip()
        tmp = float(tmp)
        tmp = tmp * 1000
        ViewCount.append(tmp)
    else:
        ViewCount.append(float(i))
```

```python
ViewCount
```

```python
CLIEN['View']
```

```python
CLIEN['ViewCount'] = ViewCount
```

```python
CLIEN.head()
```

```python
CLIEN.sort_values(by=['ViewCount'], ascending=False) # 내림차순
```

```python
KeywordRE = []
for i in range(0,len(WOS)):
    KeywordSplit = WOS['keyword'][i].split(',')
    TmpKeyword = []
    for i in KeywordSplit:
        if "''" not in i:
            TmpKeyword.append(i)
    KeywordRE.append(TmpKeyword)
```

```python
KeywordRE[0]
```

```python
len(KeywordRE[0])
```

```python
for i in KeywordRE[0]:
    print(i)
```

```python
WOS['KeywordRE'] = KeywordRE
```

```python
WOS.head(3)
```

```python
Words = []
for i in range(0, len(WOS)):
    for j in list(WOS['KeywordRE'][i]):
        Words.append(j)
```

```python
from itertools import groupby
```
