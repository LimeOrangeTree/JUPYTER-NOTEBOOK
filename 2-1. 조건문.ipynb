### for, else
- 모든 반복문 for을 진행하고 마지막에 else 뒤에 나오는 항목을 실행

```python
# for ~ else
tmp_list = [1,2,3,4,5]
for i in tmp_list:
    print(i)
else:
    print('Good!')
```

### for, if

```python
# for ~ break
tmp_list = [1,2,3,4,5]
for i in tmp_list:
    print(i)
    if i>=3:
        print('3이 나타났다!')
```

### for, break

```python
# for ~ break
tmp_list = [1,2,3,4,5]
for i in tmp_list:
    print(i)
    if i>=3:
        break
```

### pass, continue
- pass : 단순히 아무것도 하지 않는 것
- continue : 뒤에 있는 명령들을 실행하지 않고 다음 순번으로 넘어감

```python
tmp_list = [1,2,3,4,5]
for i in tmp_list:
    if i == 3:
        pass # 실행할 것이 없음
        print('this is ', i)
        print('pass', i)
    else:
        print(i)
```

```python
tmp_list = [1,2,3,4,5]
for i in tmp_list:
    if i == 3:
        continue # 다음 for loop의 순번으로 넘어감
        print(i)
        print('continue', i)
    else:
        print(i)
```

# while 문
<li>특정 조건을 만족하는 경우 지속적으로 반복을 수행</li>
- for 문의 경우, 한정된 반복 가능한 객체를 이용해서 반복 작업 수행

<strong>(while 문 예시)</strong>

while 조건:  
반복 실행 코드  
...  
(indentation) continue # while 구문 처음으로 이동하여 반복문 계속 
<br></br>
...  
(indentation) break # while 구문 탈출
<br></br>



### while 예시

```python
x = 0
while x<10:
    print('Input x is :', x)
    x = x+1
    print(x)
```

```python
x = 10

while x==10:
    x = x+1
    print(x)
```

```python
x = 10

while x==11:
    x = x+1
    print(x) # nothing happens
```

```python
x = 0

while x<10:
    x = x+1
    print(x)
    if x >7:
        break
```

```python
# # 무한 루프
# while True:
#     print(1)
```

### while, continue, break

```python
# while continue에 대한 설명 추가

x = 0

while x<10:
    print('input value:', x)
    x = x+1
    if x < 3:
        continue
    print(x)
    if x > 7:
        break
print('While Finish!')
```
